```python
from sklearn.model_selection import train_test_split
import tpot
import pandas as pd
import tqdm
from IPython.display import display
```

```python
data = pd.read_csv("datasets/dataset_adv.csv")
data = data.drop("match_id", axis=1)
display(data.head())
```

```python
X = data.drop("rad_won", axis=1).values
y = data['rad_won'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
model = tpot.TPOTClassifier(generations=15, population_size=50, verbosity=2, random_state=42)
model.fit(X_train, y_train)
```

```python
print(model.score(X_test, y_test))
```

```python
model.export('tpot_exported_pipeline_adv.py')
```
